```python
from nsepy import get_history
from datetime import datetime
import matplotlib.pyplot as plt
plt.style.use('seaborn')
tamo = get_history(symbol ='TATAMOTORS', start =datetime(2021,1,1), end = datetime(2021,7,9))
plt.figure(figsize=(14,6))
tamo['Close'].plot()
```

```python
tamofut = get_history(symbol ='TATAMOTORS', start =datetime(2021,1,1), end = datetime(2021,7,9), futures=True, expiry_date = datetime(2021,7,29))
plt.figure(figsize=(14,6))
tamofut['Close'].plot()
tamo['Close'].plot()
```
