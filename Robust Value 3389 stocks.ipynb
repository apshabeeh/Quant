```python
import pandas as pd
```

```python
df = pd.read_excel('F:\stock market\Quant\Jupiter\\3899.xlsx')
df = df[['NSE Code', 'Current Price','Market Capitalization', 'Sales', 'Price to Earning', 'Price to book value', 'EVEBITDA']]
df['Price to Sales'] = df['Market Capitalization']/ df['Sales']
df = df[['NSE Code', 'Current Price', 'Price to Sales', 'Price to Earning', 'Price to book value', 'EVEBITDA']]
df = df.dropna()
df
```

```python
rv_columns = [
    'Ticker',
    'Price',
    'Price-to-Earnings Ratio',
    'PE Percentile',
    'Price-to-Book Ratio',
    'PB Percentile',
    'Price-to-Sales Ratio',
    'PS Percentile',
    'EV/EBITDA',
    'EV/EBITDA Percentile',
    'RV Score'
]

rv_dataframe = pd.DataFrame(columns = rv_columns)
rv_dataframe[['Ticker', 'Price', 'Price-to-Earnings Ratio', 'Price-to-Book Ratio', 'Price-to-Sales Ratio', 'EV/EBITDA']] = df[['NSE Code', 'Current Price', 'Price to Earning', 'Price to book value', 'Price to Sales', 'EVEBITDA']]

from scipy import stats
metrics = {
            'Price-to-Earnings Ratio': 'PE Percentile',
            'Price-to-Book Ratio':'PB Percentile',
            'Price-to-Sales Ratio': 'PS Percentile',
            'EV/EBITDA':'EV/EBITDA Percentile'
}

for row in rv_dataframe.index:
  for metric in metrics.keys():
    rv_dataframe.loc[row, metrics[metric]] = stats.percentileofscore(rv_dataframe[metric],rv_dataframe.loc[row, metric])/100
    
rv_dataframe
```

```python
from statistics import mean

for row in rv_dataframe.index:
  value_percentiles = []
  for metric in metrics.keys():
    value_percentiles.append(rv_dataframe.loc[row, metrics[metric]])
  rv_dataframe.loc[row, 'RV Score'] = mean(value_percentiles)

rv_dataframe
```

```python
rv_dataframe.sort_values(by='RV Score', inplace = True)
# rv_dataframe = rv_dataframe[:]
# rv_dataframe.reset_index(drop=True, inplace=True)
# 
# rv_dataframe.head(30)
rv_dataframe.to_csv('Robust Value 3389 stocks.csv')
```
