```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import talib as ta
```

```python
df = yf.download('BAJFINANCE.NS', '2009-01-01', '2021-06-18')
df
```

```python
df['ema50'] = ta.EMA(df['Close'], timeperiod = 50)
df['ema200'] = ta.EMA(df['Close'], timeperiod = 200)
df = df.dropna()
df
```

```python
%matplotlib inline

plt.figure(figsize = (18,8))
plt.style.use('seaborn')
plt.plot(df['Close'], label = 'Close', color = 'b')
plt.plot(df['ema200'], label = 'EMA 200', color= 'g')
plt.plot(df['ema50'], label = 'EMA 50', color='r')
plt.legend()
plt.show()
df
```

```python
df['positions_long'] = np.NaN
for x in range (len(df)):
    if df.ema50[x] > df.ema200[x]:
        df.positions_long[x] = 1
    if df.ema50[x] < df.ema200[x]:
        df.positions_long[x] = 0
df['positions_long'] = df['positions_long'].fillna(method='ffill')
df['price_difference'] = df['Close'] - df['Close'].shift(1)
df['pnl'] = df['positions_long'].shift(1)*df['price_difference']
df['cum_pnl'] = df['pnl'].cumsum()
df
```

```python
df['rolling_max'] = df['cum_pnl'].rolling(250, min_periods=1).max()
df['daily_drawdown'] = df['cum_pnl']-df['rolling_max']
df['max_daily_drawdown'] = df['daily_drawdown'].rolling(250, min_periods=1).max()
fig, axs = plt.subplots(2, figsize = (18,8))
axs[0].plot(df['max_daily_drawdown'])
axs[1].plot(df['cum_pnl'])
```
